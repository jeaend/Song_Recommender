### Objective

Using web scraping to retrieve a list of weekly top 100 songs. 
  
*Output: 1_Top_100.csv, should have columns: 'title', artist*  

Popvortex maintains a weekly Top 100 of "hot" songs: [http://www.popvortex.com/music/charts/top-100-songs.php](http://www.popvortex.com/music/charts/top-100-songs.php).

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

> retrieve data

```python
url = "https://www.popvortex.com/music/charts/top-100-songs.php"
response = requests.get(url)
# check response is correst
response.status_code 
```

```python
# check content 
response.content
```

> create soup

```python
soup = BeautifulSoup(response.content, "html.parser")
```

> retrieve songs

```python
artist_paragraph = soup.select("body > div.container > div:nth-child(4) > div.col-xs-12.col-md-8 > div.chart-wrapper > div.feed-item > div.chart-content > p.title-artist")
#artist_paragraph
```

> creates a list of tupels with (artist, title)

```python
title_artist_list = []
for p in artist_paragraph:
    artist_element = p.select('em.artist')
    title_element = p.select('cite.title')
    if title_element and artist_element:
        title = title_element[0].get_text()  
        artist = artist_element[0].get_text() 
        title_artist_list.append((artist, title))


# sanity check
len(title_artist_list)
#title_artist_list
```

> to dataframe

```python
df = pd.DataFrame(title_artist_list, columns=['artist', 'title'])
df.to_csv('./cache/1_Top_100.csv', index=False)
```
